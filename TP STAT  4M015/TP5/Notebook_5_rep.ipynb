## Question 3

```r
mod <- lm(maxO3~., data = oz)
summary(mod)
```

Le test de Fisher global est significatif (la p-value est quasi nulle), donc on en conclut qu'il y a au moins une
variable qui a un coefficient non nul.

Le test de nullit des coefficients n'est significatif que pour les variables **Ne9** et **maxO3v**. Pour toutes les autres variables, la p-value du test est leve (entre 0.12 et .98), ce qui veut dire qu'on ne peut pas rejeter l'hypothse de nullite du coefficient. Or, chacun de ces tests est effectu dans le modle complet avec 10 variables explicatives (+ la constante). La conclusion des tests avec une p-value leve est : en prsence de toutes les autres variables explicatives, la variable j (pour laquelle on teste $\beta_j=0$) n'est pas significative.

Pour tester la nullit des 8 variables qui ont une grande p-value ici, il faudrait faire un test de Fisher qui teste la nullit simultane de ces 8 coefficients.

## Question 4

```r
residus <- rstudent(mod)
n <- length(oz$maxO3)
plot(1:n, residus, pch = 16, xlab = 'Index', ylab = 'Residus studentises',
     main = 'Valeurs aberrantes')
abline(-2, 0, lty = 2)
abline(2, 0, lty = 2)
IDval.ab <- (1:n)[abs(residus)>2]
text(IDval.ab, residus[IDval.ab], IDval.ab, pos = 4, col = 'blue')
```

Il n'y a que 4 observations en dehors de l'intervalle $[-2,2]$, ce qui est acceptable pour un jeu de donnes de taille 112. De plus, ces rsidus ne sont pas trs loigns des limites d'intervalle. Ce ne sont pas de vritables observations aberrantes, elles sont juste un peu limites sauf peut-tre l'observation 58.

D'aprs le thorme du cours, si les residus $\varepsilon_i$ suivent la loi normale $\mathcal N(0,\sigma^2)$, alors les rsidus studentiss suivent la loi de Student $\mathcal T(n-p-1)$. Commenons par calculer les quantiles d'ordre $i/n$ de la loi $\mathcal T(n-p-1)$, puis traons un QQ-plot.

```r
quant.t <- qt((1:n)/n,n-3)
plot(quant.t, sort(residus), xlab = 'Student T(n-p-1)',
     ylab = 'Quantiles empiriques des rsidus', main = 'QQ-plot des rsidus')
abline(0, 1, col ='blue')
```

L'alignement des points sur la premire bissectrice est presque parfaite sauf deux points pour la queue  droite et un pour la queue  gauche, ce qui confirme l'hypothse selon laquelle les residus thoriques $\varepsilon_i$ suivent la loi normale $\mathcal N(0,sigma^2)$.

```r
residus.std <- rstandard(mod)
plot(1:n, residus.std, pch = 4, col = 'blue', main = 'Residus', xlab = 'Index',
     ylab = 'Residus')
points(1:n, residus, col = 'red')
legend(90, 2.5 , c('standardise', 'studentise'), col = c('blue', 'red'), pch = c(4, 1))
```

Nous voyons que pour la grande majorit des points les deux rsidus sont quasi identiques. Des lgres diffrences n'apparaissent que pour les rsidus dont la valeur abolue est relativement leve. Dans ce cas, les valeurs des rsidus studentiss (en valeur absolue) sont plus leves.

```r
levier <- hatvalues(mod)
plot(1:n, levier, xlab = 'Index', ylab = 'Poids h_ii', main = 'Points leviers')
p <- mod$rank
seuil1 <- 2*p/n
seuil2 <- 3*p/n
abline(seuil1, 0, lty = 2)
abline(seuil2, 0, lty = 3)
```

Aucun point ne dpasse le deuxime seuil, autrement dit aucun point n'est levier. 

```r
cook <- cooks.distance(mod)
plot(1:n, cook, xlab = 'Index', ylab = 'Distance de Cook', main = 'Distance de Cook',
     ylim = c(0,.15))
s1 <- qf(0.5, p, n-p)
s2 <- qf(0.1, p, n-p)
abline(s2, 0, lty = 2)
abline(s1, 0, lty = 3)
```

Aucun point ne dpasse le premier seuil, ainsi aucun point n'est pas aberrant au sens de la distance de Cook. Nous ne supprimons donc aucun point. 

```r
resid.st <- rstudent(mod)
nom_var <- names(oz)
n <- nrow(oz)
val_aber <- (1:n)[abs(resid.st) > 2]  # indices des valeurs aberrantes (pour les signaler dans les graphiques)
par(mfrow = c(2, 2))
for (i in 1:4){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}

for (i in 5:8){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}

par(mfrow = c(1,2))
for (i in 9:10){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}
```

Il ne semble pas y avoir de vritable tendance dans les graphiques des rsidus studentiss en fonction des variables explicatives, surtout si on enlve les 4 valeurs aberrantes


```r
par(mfrow = c(1, 2))
plot(resid.st, xlab = 'Indice', pch = 16)
points(val_aber, resid.st[val_aber], pch = 4, cex = 2)
plot(resid.st~mod$fitted.values, xlab = 'Fitted values', pch = 16)
points(mod$fitted.values[val_aber], resid.st[val_aber], pch = 4, cex = 2)
```

```r
library(leaps)
```

 - Pour le critre $R^2$

```r
choix <- regsubsets(maxO3~., data = oz, nvmax = 11)
plot(choix, scale = "r2")
```

On cherche donc le modle qui maximise le coefficient de dtermination $R^2$. On retrouve le fait que le coefficient $R^2$ augmente avec le nombre de variables et que pour une comparaison entre modles embots, il n'est pas judicieux de considrer ce critre.
 - Pour le critre $R^2_a$


```r
plot(choix, scale = "adjr2")
```

Le but est de maximiser le coefficient $R^2_a$. Ici, c'est le modle contenant la constante, **T12**, **N9**, **Vx9** et **maxO3v**. 

 - Pour le critre BIC

```r
plot(choix, scale = "bic")
```

On cherche  minimiser le critre BIC. On choisit donc le modle contenant la constante, la variable **T12**, **Ne9**, **Vx9** et **maxO3v**. On retrouve les mmes variables que celles obtenues avec le critre $R^2_a$.

Effectuons la rgression linaire avec ces variables


```r
reg <- lm(maxO3~T12+Ne9+Vx9+maxO3v, data = oz)
summary(reg)
```
