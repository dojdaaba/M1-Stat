## Question 3

```r
mod <- lm(maxO3~., data = oz)
summary(mod)
```

Le test de Fisher global est significatif (la p-value est quasi nulle), donc on en conclut qu'il y a au moins une
variable qui a un coefficient non nul.

Le test de nullité des coefficients n'est significatif que pour les variables **Ne9** et **maxO3v**. Pour toutes les autres variables, la p-value du test est élevée (entre 0.12 et .98), ce qui veut dire qu'on ne peut pas rejeter l'hypothèse de nullite du coefficient. Or, chacun de ces tests est effectué dans le modèle complet avec 10 variables explicatives (+ la constante). La conclusion des tests avec une p-value élevée est : en présence de toutes les autres variables explicatives, la variable j (pour laquelle on teste $\beta_j=0$) n'est pas significative.

Pour tester la nullité des 8 variables qui ont une grande p-value ici, il faudrait faire un test de Fisher qui teste la nullité simultanée de ces 8 coefficients.

## Question 4

```r
residus <- rstudent(mod)
n <- length(oz$maxO3)
plot(1:n, residus, pch = 16, xlab = 'Index', ylab = 'Residus studentises',
     main = 'Valeurs aberrantes')
abline(-2, 0, lty = 2)
abline(2, 0, lty = 2)
IDval.ab <- (1:n)[abs(residus)>2]
text(IDval.ab, residus[IDval.ab], IDval.ab, pos = 4, col = 'blue')
```

Il n'y a que 4 observations en dehors de l'intervalle $[-2,2]$, ce qui est acceptable pour un jeu de données de taille 112. De plus, ces résidus ne sont pas très éloignés des limites d'intervalle. Ce ne sont pas de véritables observations aberrantes, elles sont juste un peu limites sauf peut-être l'observation 58.

D'après le théorème du cours, si les residus $\varepsilon_i$ suivent la loi normale $\mathcal N(0,\sigma^2)$, alors les résidus studentisés suivent la loi de Student $\mathcal T(n-p-1)$. Commençons par calculer les quantiles d'ordre $i/n$ de la loi $\mathcal T(n-p-1)$, puis traçons un QQ-plot.

```r
quant.t <- qt((1:n)/n,n-3)
plot(quant.t, sort(residus), xlab = 'Student T(n-p-1)',
     ylab = 'Quantiles empiriques des résidus', main = 'QQ-plot des résidus')
abline(0, 1, col ='blue')
```

L'alignement des points sur la première bissectrice est presque parfaite sauf deux points pour la queue à droite et un pour la queue à gauche, ce qui confirme l'hypothèse selon laquelle les residus théoriques $\varepsilon_i$ suivent la loi normale $\mathcal N(0,sigma^2)$.

```r
residus.std <- rstandard(mod)
plot(1:n, residus.std, pch = 4, col = 'blue', main = 'Residus', xlab = 'Index',
     ylab = 'Residus')
points(1:n, residus, col = 'red')
legend(90, 2.5 , c('standardise', 'studentise'), col = c('blue', 'red'), pch = c(4, 1))
```

Nous voyons que pour la grande majorité des points les deux résidus sont quasi identiques. Des légéres différences n'apparaissent que pour les résidus dont la valeur abolue est relativement élevée. Dans ce cas, les valeurs des résidus studentisés (en valeur absolue) sont plus élevées.

```r
levier <- hatvalues(mod)
plot(1:n, levier, xlab = 'Index', ylab = 'Poids h_ii', main = 'Points leviers')
p <- mod$rank
seuil1 <- 2*p/n
seuil2 <- 3*p/n
abline(seuil1, 0, lty = 2)
abline(seuil2, 0, lty = 3)
```

Aucun point ne dépasse le deuxième seuil, autrement dit aucun point n'est levier. 

```r
cook <- cooks.distance(mod)
plot(1:n, cook, xlab = 'Index', ylab = 'Distance de Cook', main = 'Distance de Cook',
     ylim = c(0,.15))
s1 <- qf(0.5, p, n-p)
s2 <- qf(0.1, p, n-p)
abline(s2, 0, lty = 2)
abline(s1, 0, lty = 3)
```

Aucun point ne dépasse le premier seuil, ainsi aucun point n'est pas aberrant au sens de la distance de Cook. Nous ne supprimons donc aucun point. 

```r
resid.st <- rstudent(mod)
nom_var <- names(oz)
n <- nrow(oz)
val_aber <- (1:n)[abs(resid.st) > 2]  # indices des valeurs aberrantes (pour les signaler dans les graphiques)
par(mfrow = c(2, 2))
for (i in 1:4){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}

for (i in 5:8){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}

par(mfrow = c(1,2))
for (i in 9:10){
  plot(resid.st~oz[, i+1], xlab = nom_var[i+1], pch = 16)
  points(oz[val_aber, i+1], resid.st[val_aber], pch = 4, cex = 2)
}
```

Il ne semble pas y avoir de véritable tendance dans les graphiques des résidus studentisés en fonction des variables explicatives, surtout si on enlève les 4 valeurs aberrantes


```r
par(mfrow = c(1, 2))
plot(resid.st, xlab = 'Indice', pch = 16)
points(val_aber, resid.st[val_aber], pch = 4, cex = 2)
plot(resid.st~mod$fitted.values, xlab = 'Fitted values', pch = 16)
points(mod$fitted.values[val_aber], resid.st[val_aber], pch = 4, cex = 2)
```

# Exercice 2

```r
library(leaps)
```

 - Pour le critère $R^2$

```r
choix <- regsubsets(maxO3~., data = oz, nvmax = 11)
plot(choix, scale = "r2")
```

On cherche donc le modèle qui maximise le coefficient de détermination $R^2$. On retrouve le fait que le coefficient $R^2$ augmente avec le nombre de variables et que pour une comparaison entre modèles emboîtés, il n'est pas judicieux de considérer ce critère.
 - Pour le critère $R^2_a$


```r
plot(choix, scale = "adjr2")
```

Le but est de maximiser le coefficient $R^2_a$. Ici, c'est le modèle contenant la constante, **T12**, **N9**, **Vx9** et **maxO3v**. 

 - Pour le critère BIC

```r
plot(choix, scale = "bic")
```

On cherche à minimiser le critère BIC. On choisit donc le modèle contenant la constante, la variable **T12**, **Ne9**, **Vx9** et **maxO3v**. On retrouve les mêmes variables que celles obtenues avec le critère $R^2_a$.

Effectuons la régression linéaire avec ces variables


```r
reg <- lm(maxO3~T12+Ne9+Vx9+maxO3v, data = oz)
summary(reg)
```
